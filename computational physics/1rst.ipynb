```python
u=np.random.rand(100)
zz=np.arcsin(2*u-1) 
plt.hist(zz)
plt.xlabel('Μετρήσεις')
plt.ylabel('Συχνότητα')
plt.title('Ιστόγραμμα Μετρήσεων')
plt.show()
```

```python
# Υπολογισμός της διαφοράς των μέσων τιμών
mean_data = np.mean(y)
mean_random = np.mean(zz)
diff_means = abs(mean_data - mean_random)

std_dev = np.std(x)
one_sigma = std_dev

# Έλεγχος αν η διαφορά των μέσων τιμών είναι μικρότερη από το 1σ
if diff_means < one_sigma:
    print("Οι μέσες τιμές είναι εντός 1σ η μία από την άλλη.")
else:
    print("Οι μέσες τιμές δεν είναι εντός 1σ η μία από την άλλη.")
```
